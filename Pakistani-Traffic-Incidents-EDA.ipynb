```python
data.rename(columns={'*- Islamabad Included.':'Place','Unnamed: 1':'Year','Unnamed: 2':'Total_accidents','Unnamed: 3':'Fatal_Accident','Unnamed: 4':'Non-Fatal_Accidents','Unnamed: 5':'Killed','Unnamed: 6':'Injured','Unnamed: 7':'Total_vehicles_involved'},inplace=True)
```

```python
data=data.iloc[2:]
data
```

```python
data['Year']=data['Year'].replace("2012-13 * ","2012-13")
data['Year']=data['Year'].replace("2013-14*","2013-14")
data['Year']=data['Year'].replace("2014-15*","2014-15")
data['Year']=data['Year'].replace("2015-16*","2015-16")
data['Year']=data['Year'].replace("2016-17*","2016-17")
```

```python
data['Year']=data['Year'].str.split("-").str[0]
data=data.explode('Year',ignore_index=True)
```

```python
data=data.astype({'Place':str,'Year':int,'Total_accidents':int,'Fatal_Accident':int,'Non-Fatal_Accidents':int,
                 'Killed':int,'Injured':int,'Total_vehicles_involved':int})
data
```

In which province the most accidents have occurred?


```python
province=data.groupby('Place')
province
len(province)
```

```python
province.size()
```

```python
province['Total_accidents'].sum().sort_values(ascending=False)
```

In which province the least accidents have occurred?

```python
province['Total_accidents'].sum().sort_values(ascending=True)
```
